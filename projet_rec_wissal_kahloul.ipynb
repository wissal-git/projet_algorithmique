<h4>1)  Formuler le problème selon le principe « Diviser pour régner » </h4>
    
le principe consiste a formuler une fonction qui place seulement une reine sur la premiere lignes du tableau ,effectuer un teste de validité , ainsi cette fonction est regenerer a chaque fois qu'on insere une nouvelle reine. 
    

<h3>                  <span style =" color:blue   "> 
    
   _stratégie récursif_ :</span>
 </h3>
    
Sachant qu’il y a des reines bien placées dans les lignes 1 à k − 1 et connaissant leur place, trouver toutes les solutions possibles en plaçant convenablement les reines de k à n. 
    
**2) Ecrire un algorithme récursif appelé placer_reine permettant de placer une reine sur un échiquier.** 
   
<h3>                  <span style =" color:blue   "> 
    
   _Algorithme de placement_ :</span>
 </h3>
Cette fonction donne la position de 8dames :

**Etape :**

1- Placer une dame sur la premiére ligne

2- Parcourir à chaque fois la listes des emplacements libres dans la prochaine ligne 

3- Placer la reine dans les cases libre en verifiant les teste de     validité par colomne et par diagonal 

L'algorithme case_libre teste si la case à la ligne et à la colonne  n'est pas menacée par les reines précédemment placées. La vérification se fait en observant qu'aucune reine précédente n'est placée sur la même ligne, sur la même colonne ou les mêmes diagonales.



<h3>  <span style =" color:orange  "> L’algorithme case_libre() est alors : </span> </h3>


**<span style =" color:green  ">Fonction case_libre (lst_dame):liste</span>**

**VARIABLES**

lst_possible,lst_valide :liste de tuple(entier,entier)

i,j:position tuple

Test_valide: booléen

**DEBUT**
  
  Test_valide← VRAI
   
   Pour i = premier(lst_possible) à dernier (lst_possible) faire
   
       Pour j = premier (lst_dame) à dernier (lst_dame) faire 
               Si ( lst_possible[1] =lst_dame[1]) alors
                             Test_valide ← faux       
               Sinon   
                  si    ( abs((i[0]-j[0])/(i[1]-j[1]))=1 ) alors 
                              Test_valide ← faux 
               finsi   
       fin pour     
       Si (test_valide<-vrai) alors
                lst_valide <- lst_valide + i
       finsi         
   fin pour   
   
   Retourner lst_valide 
   
**FIN**

```python
def case_libre(lst_dame):
    lst_possible =[(len(lst_dame), i) for i in range(8)]
    
    lst_valide =[]
    for i in lst_possible :
        test_valide=True 
        for j in lst_dame:
            if i[1]==j[1]:
                test_valide =False 
            elif abs((i[0]-j[0])/(i[1]-j[1]))==1 :
                test_valide=False
        if test_valide:
            lst_valide.append(i)
    return lst_valide  

```

**3) Ecrire un algorithme récursif appelé "placer_toute_reine" permettant de placer toutes les reines sur un échiquier. Votre algorithme devra chercher au moins une solution possible.**    

L'algorithme Placer_toute_Reine est chargé de placer les reines les unes après les autres . l'algorithme ne s'arrête pas 
dès que les  reines sont placées mais poursuit son investigation dans l'arbre jusqu'au bout. Les n alternatives de placement d'une reine sur une ligne se concrétisent dans l'algorithme par une boucle suivie d'un test pour  vérifier que la 
case candidate n'est pas menacée. 

    
<h3>  <span style =" color:orange  ">     algorithme placer_toute_reines :</span> </h3>
    
**<span style =" color:green ">     Fonction placer_toute_reine (lst_dame):liste </span>**

**VARIABLES**
 
i :position

**DEBUT**
 
   Pour i =premier(lst_dame) à dernier (lst_dame) faire 
   
                Si ( length(lst_dame)=7) alors
                           lst_dame<-lst_dame+ i
                Sinon
                   lst_dame<-placer_toute_reine(lst_dame+i)
                finsi
         fin pour     
    Retourner lst_dame
**FIN**



```python

def   placer_toute_reine(lst_dame=[]):
    for i in case_libre(lst_dame):
        #si la long liste des dames est 7 encore trouver un emplacement libre 
        if len(lst_dame)==7:
            #imprimer solution liste des dames en lui ajoutant l'esspace libre
            print(lst_dame + [i])
        else :
            #ajouter la liste des dames qu'on viennt de trouver 
            placer_toute_reine(lst_dame +[i])
            
            
print( placer_toute_reine())
```
